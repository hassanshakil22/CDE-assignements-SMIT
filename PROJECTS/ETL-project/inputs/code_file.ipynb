```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd
from selenium.webdriver.common.by import By  
from io import StringIO
import lxml
import html5lib 
```

```python
# 
path = r"C:\Users\hhhh\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe"
driver = webdriver.Chrome(path)
def extract(url):
    driver.get(url)
    table  = driver.find_element(By.CLASS_NAME, "wikitable.sortable.mw-collapsible.jquery-tablesorter.mw-made-collapsible")
    table_head = table.find_element(By.TAG_NAME, "thead")
    table_headers = []
    data = []

    for th in table_head.find_elements(By.TAG_NAME , "th"):
        table_headers.append(th.text)
    table_headers[2] = "Market cap (US$ billion)"    
    table_body = table.find_element(By.TAG_NAME, "tbody")

    for tr in table_body.find_elements(By.TAG_NAME, "tr"):
        tds =  tr.find_elements(By.TAG_NAME, "td")

        data.append( {
            table_headers[0] :  tds[0].text,
            table_headers[1] : tds[1].text,
            table_headers[2] : tds[2].text
        })

    df = pd.DataFrame(data,)
    return df

df = extract("https://web.archive.org/web/20230908091635/https://en.wikipedia.org/wiki/List_of_largest_banks")

```

```python
#transform 
print(df)
def transform (df , csv ):
    rates_df = pd.read_csv(csv)
    print(rates_df)
    rates_dict = rates_df.set_index('currency')['rate'].to_dict()
    print(rates_dict)
    for currency, rate in rates_dict.items():
        df[f'price_in_{currency.upper()}'] = pd.to_numeric(df['Market cap (US$ billion)']) * rate

    return df  

transfromed_df = transform(df, 'currency_rates.csv')
print(transfromed_df)
```
