```python
url = "https://m.timesjobs.com/mobile/jobs-search-result.html?txtKeywords=flutter&cboWorkExp1=-1&txtLocation="

web = requests.get(url).content
web

```

```python
soup = BeautifulSoup(web,"html.parser")
soup
```

```python
titles = soup.find_all('div',class_='srp-job-heading')
# print(titles)
for title in titles:
    # print (title.)
    h3 = title.find('h3')
    h4 = title.find('h4')
    print(h3.text)
    print(h4.text)
```

```python
joblist = []
jobs = soup.find_all('div',class_='srp-job-heading')
for job in jobs:
    title= job.find('h3')
    
    sub = job.find('h4')
    company = sub.find('span','srp-comp-name')
    posting_time = sub.find('span','posting-time')
    # print(title.text)
    # print(company.text)
    # print(posting_time.text)
    jobs = {
        'job-title' : title.text,
        'company-name' : company.text,
        'posting time' : posting_time.text,
    }
    joblist.append(jobs)
for job in joblist:
    print(joblist)
    
```

```python
import pandas as pd

df = pd.DataFrame(joblist)
print(df)
```
