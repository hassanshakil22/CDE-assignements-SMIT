```python
import numpy as np 
import matplotlib.pyplot as plt 
from datetime import datetime,timedelta
```

```python
start_date = datetime(2024,1,1)
end_date = datetime(2024,12,31)

num_of_days = (end_date - start_date).days
print(num_of_days)
```

```python
date_list = []
for i in range(num_of_days):
    date_list.append(start_date.date())
    start_date += timedelta(days=1)

print(date_list)
```

```python
np.random.seed(0)
stock_price_list1 = (np.random.randn(num_of_days).cumsum() * 50).tolist()
stock_price_list2 = (np.random.randn(num_of_days).cumsum() * 50).tolist()
print(stock_price_list1)
```

```python
plt.figure(figsize= (10,10))
plt.plot(date_list,stock_price_list1,marker=".",color="g",label = "apple")
plt.title("Stock Analysis")
plt.xlabel("Dates")
plt.ylabel("price")




plt.show
```

```python
plt.figure(figsize= (10,10))
plt.plot(date_list,stock_price_list1,marker=".",color="g",label = "apple")
plt.plot(date_list,stock_price_list2,marker="x",color="b",label = "microsoft")
plt.title("Stock Analysis")
plt.xlabel("Dates")
plt.ylabel("price")




plt.show
```
